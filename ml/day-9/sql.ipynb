1. Data Definition Language

```python
import sqlite3

# Connect to database (creates file if it doesn't exist)
conn = sqlite3.connect("company.db")

# Create cursor
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT,
    salary REAL
)
""")

# Save changes
conn.commit()

print("Table created successfully!")

# Close connection
conn.close()
```

5. Transaction Control Language

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary REAL
)
""")

# Insert data
cursor.execute("INSERT INTO Employees (name, salary) VALUES (?, ?)", ('Ravi', 30000))

# Save changes permanently
conn.commit()

print("Changes committed successfully!")

# Close connection
conn.close()
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary REAL
)
""")

# Start transaction (default behavior)
try:
    # Insert data
    cursor.execute("INSERT INTO Employees (name, salary) VALUES (?, ?)", ('Ravi', 30000))

    # Simulate error
    x = 10 / 0   # This will cause an error

    # Commit if everything is fine
    conn.commit()

except:
    print("Error occurred! Rolling back changes...")
    conn.rollback()   # Undo changes

finally:
    conn.close()
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary REAL
)
""")

try:
    # Insert first record
    cursor.execute("INSERT INTO Employees (name, salary) VALUES (?, ?)", ('Ravi', 30000))

    # Create savepoint
    cursor.execute("SAVEPOINT sp1")

    # Insert second record
    cursor.execute("INSERT INTO Employees (name, salary) VALUES (?, ?)", ('Rani', 25000))

    # Something goes wrong → rollback to savepoint
    cursor.execute("ROLLBACK TO sp1")

    # Commit remaining changes
    conn.commit()

    print("Rollback to savepoint done!")

except:
    conn.rollback()
    print("Full rollback executed!")

finally:
    conn.close()
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Execute query
cursor.execute("SELECT FirstName FROM Customer")

# Fetch results
rows = cursor.fetchall()

# Print results
for row in rows:
    print(row)

# Close connection
conn.close()
```

Example 4: SELECT with GROUP BY Clause

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Execute GROUP BY query
cursor.execute("""
SELECT Country, COUNT(*) AS customer_count
FROM Customer
GROUP BY Country
""")

# Fetch results
rows = cursor.fetchall()

# Display results
for row in rows:
    print("Country:", row[0], "| Customers:", row[1])

# Close connection
conn.close()
```

Example 5: SELECT with DISTINCT Clause


```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Execute DISTINCT query
cursor.execute("SELECT DISTINCT Country FROM Customer")

# Fetch results
rows = cursor.fetchall()

# Display results
for row in rows:
    print(row[0])

# Close connection
conn.close()
```

Example 6: SELECT Statement with HAVING Clause

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Execute query
cursor.execute("""
SELECT Country, COUNT(*) AS customer_count
FROM Customer
GROUP BY Country
HAVING COUNT(*) >= 2
""")

# Fetch results
rows = cursor.fetchall()

# Display results
for country, count in rows:
    print(f"Country: {country} | Customers: {count}")

# Close connection
conn.close()
```

Example 7: SELECT Statement with ORDER BY clause
